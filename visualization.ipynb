```python
import cv2
import pandas as pd
from PIL import Image
```

```python
train_labels = pd.read_csv('data/train_labels.csv')
```

```python
train_labels.head()
```

```python
def draw_boxes(image_name):
    selected_value = train_labels[train_labels.filename == image_name]
    img = cv2.imread('raw_images/{}'.format(image_name))
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    for index, row in selected_value.iterrows():
        img = cv2.rectangle(img, (row['xmin'], row['ymin']), (row['xmax'], row['ymax']), (0, 255, 0), 3)
    return img

```

```python
Image.fromarray(draw_boxes('BLF1306.jpg'))
```

```python
Image.fromarray(draw_boxes('BLF1084.jpg'))
```
